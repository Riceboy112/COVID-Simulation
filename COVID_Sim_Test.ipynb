```python
import matplotlib.pyplot as plt
import random
import pandas as pd
import math
import matplotlib.pyplot as plt
from PIL import Image
import glob

global file_number, helpa, log_file, df, fig, Stat, Day, Moverslist, YesterdayPatients, Distlimit

# Generates a random patient zero that will spread the virus.
def Patient_Zero(xlimit, ylimit):
    x = random.uniform(0, xlimit)
    y = random.uniform(0, ylimit)
    return x, y

# Generates a list of humans that will be seperated into 5 different categories.
def Generate(GroupSize, xlimit, ylimit):
    df = pd.DataFrame(columns = 'X,Y,Covid-19,Day'.split(','))
    
    for i in range(GroupSize):
        df.loc[i, 'X'], df.loc[i, 'Y'] = Patient_Zero(xlimit, ylimit)
        df.loc[i, 'Covid-19'] = False
    
    samplesize = math.floor(GroupSize/100)
    MoversList = df.sample(n = samplesize).index.values.tolist()
    
    StatofDay = pd.DataFrame(columns='Not Infected,Infected,Hospitalized,Cured,Dead'.split(',')) ##################HERERERERERRERERERRERER
    return df, StatofDay, MoversList
    
# Color codes the human simulation key. Makes it easier to distinguish and visually appealing
def Human_Simulation_Color(df):
    colors = []
    for l in df.index:
        if (df.loc[l, 'Covid-19'] == True): #Infected
            colors.append('red')
        elif (df.loc[l, 'Covid-19'] == 666): #Dead
            colors.append('black')
        elif (df.loc[l, 'Covid-19'] == 115): #Hospitalized
            colors.append('purple')
        elif (df.loc[l, 'Covid-19'] == 7): #Cured
            colors.append('green')
        else:
            colors.append('blue') #Healthy
    return colors

# Color codes the line graph's key. Makes it easier to distinguish and visually appealing
def Line_Graph_Color(Stat):
    colors = []
    for i in Stat.columns:
        if (i == 'Infected'): #Infected
            colors.append('red')
        elif (i == 'Dead'): #Dead
            colors.append('black')
        elif (i == 'Hospitalized'): #Hospitalized
            colors.append('purple')
        elif (i == 'Cured'): #Cured
            colors.append('green')
        else:
            colors.append('blue') #Not Infected
    return colors

# Plots the line graph and human simulation graph using the data gathered from the simulation
def Plot():
    #global df, fig, Stat, Day, Moverslist
    colors = Human_Simulation_Color(df)
    ld = ['Not Infected', 'Infected', 'Hospitalized', 'Cured', 'Death Toll']
    axs[0].cla()
    axs[0].scatter(df['X'], df['Y'], s = 1, c = colors)
    for i in MoversList:
        axs[0].scatter(df.loc[i, 'X'], df.loc[i, 'Y'], s = 6, facecolors = 'none', edgecolors = 'black')
        axs[0].text(df.loc[i, 'X'] + 0.02, df.loc[i, 'Y'] + 0.02, str(i), fontsize = 5)
    colors = Line_Graph_Color(Stat)
    single_digit_day = str(Day)
    title = 'Day' + single_digit_day
    axs[0].set_title(title, loc = 'left')
    axs[0].set_yticklabels([])
    axs[0].set_xticklabels([])
    axs[0].tick_params(
    which ='both',      # Both major and minor ticks are affected
    bottom = False,      # Ticks along the bottom edge are off
    top = False,         # Ticks along the top edge are off
    right = False,      # Ticks along the right edge are off
    left = False,         # Ticks along the left edge are off
    labelbottom = False) # Labels along the bottom edge are off
    
    # Creates the plot key using the 5 different criteria
    axs[1].cla()
    axs[1].plot(Stat['Not Infected'], label = ld[0], color = colors[0])
    axs[1].plot(Stat.Infected, label = ld[1], color = colors[1])
    axs[1].plot(Stat.Hospitalized, label = ld[2], color = colors[2])
    axs[1].plot(Stat.Cured, label = ld[3], color = colors[3])
    axs[1].plot(Stat.Dead, label = ld[4], color = colors[4])
    axs[1].legend(bbox_to_anchor = (0,1), loc = 'upper left', borderaxespad = 0.)
    plt.xlabel('Days')
    
    # Inserts a 0 in front of any days that is under 10
    if Day < 10: 
        single_digit_day = '0' + single_digit_day
    title = 'Day ' + single_digit_day + '.png'
    plt.savefig(title)
    plt.show()
    
    return

# From all the snapshots of the graph, take all the photos and turn it into a gif
def create_gif(): 
    # Create frames
    frames = []
    imgs = glob.glob("*.png")
    for i in imgs:
        new_frame = Image.open(i)
        frames.append(new_frame)
        
    # Save into GIF
    frames[0].save('COVID_Simulation_' + str(file_number) + '.gif', format = 'GIF',
          append_images=frames[1:],
          save_all = True,
          duration = 500, loop = 0)

# Simulates a person getting infected.
def infect(Person):
    #global df, Day
    if random.random() > 0.25 and Day > 3 :
        return
    if df.loc[Person, 'Covid-19'] == False:
        df.loc[Person, 'Covid-19'], df.loc[Person, 'Day'] = True, Day
    
# Simulates human daily lives and how they move around
def Moving_Humans(xlimit, ylimit):
    #global df, MoversList
    for i in MoversList:
        if (df.loc[i, 'Covid-19'] == 115) or (df.loc[i, 'Covid-19'] == 666): 
            MoversList.remove(i)
        df.loc[i, 'X'], df.loc[i, 'Y'] = (df.loc[i, 'X'] + random.uniform(1, xlimit/3)) % xlimit, (df.loc[i,'Y'] + random.uniform(1, ylimit/3)) % ylimit

# Checks to see if the humans are infected if they are within the distance to get infected.
def check(i, j):
    #global df, YesterdayPatients, Distlimit
    Dist = math.sqrt((df.loc[i, 'X'] - df.loc[j, 'X']) ** 2 + (df.loc[i, 'Y'] - df.loc[j, 'Y']) ** 2)
    flag = ((YesterdayPatients[i] == True) ^ (YesterdayPatients[j] == True)) and Dist < Distlimit
    return flag
    
# Simulates human interactions with each other. Humans may get infected or not. 
def Interaction():
    #global Day, df
    for i in range(len(df)):
        for j in range(i):
            if check(i,j):
                if (df.loc[i, 'Covid-19'] == False):
                    infect(i)
                else:
                    infect(j)

# Simulates humans death due to COVID.
def kill():
    samplesize = math.floor(len(df[df['Covid-19'] == True]) * .005 + len(df[df['Covid-19'] == 115]) * .005)
    if samplesize > len(df[df['Covid-19'] == True]):
        return
    df.loc[df[df['Covid-19'] == True].sample(n = samplesize).index.values.tolist(), 'Covid-19'] = 666
    return

# Simulates infected humans choosing whether to go to the hospital or not.
def hospitilize():
    samplesize = math.floor(len(df[df['Covid-19'] == True]) * 0.03)
    if samplesize > len(df[df['Covid-19'] == True]): 
        return
    df.loc[df[df['Covid-19'] == True].sample(n = samplesize).index.values.tolist(), 'Covid-19'] = 115
    return

# Simulates an infected human getting cured from COVID
def Cured():
    df.loc[(df['Day'] < Day - 10) & (df['Covid-19'] == True) , 'Covid-19'] = 7
    df.loc[(df['Day'] < Day - 21) & (df['Covid-19'] == 115) , 'Covid-19'] = 7 
    return
    
# Simulates human daily lives and how COVID affects infected and not infected humans.
def Tomorrow(): # To Be checked and Resolved!!!
    global Day
    Day += 1
    kill()
    hospitilize()
    Cured()
    Moving_Humans(xlimit, ylimit)
    Interaction()
    
# Simulates each passing day. Will update the status of the humans for each day.
def Count(Day):
    List = list(df['Covid-19'])
    Stat.loc[Day,'Not Infected'] = List.count(False)
    Stat.loc[Day,'Infected'] = List.count(True)    
    Stat.loc[Day,'Hospitalized'] = List.count(115)    
    Stat.loc[Day,'Cured'] = List.count(7)
    Stat.loc[Day,'Dead'] = List.count(666) 
    return
    
# Keeps track of each day and logs it into a file.
def write_log(*args):
    #global log_file
    line = ' '.join([str(a) for a in args])
    log_file.write(line + '\n')
    print(line)
    
# Main ------------------------------------------------------------------- below

# Gets the user to input data to start the COVID simulation

print("Time references for inputted population below.")
print("Any population above 5000 will take over 4.5 hours to simulate.")
print("Any population above 2000 will take over 2 hours to simulate.")
print("Population at 1250 will take about 35 minutes to simulate.")
print("Population at 1000 will take about 15 minutes to simulate.")

while(True):
    n = int(input("Enter a population number: ")) # Enter a population number
    xlimit = int(input("Enter the horizontal length (in feet): "))
    ylimit = int(input("Enter the vertical length (in feet): "))
    enclosed_square = (xlimit * ylimit)
    file_number = n
    Distlimit = float(input("Enter the max distance (in feet) required for contamination: ")) # Enter max distance
    if(n > enclosed_square):
        print("\n\nError!!! The area must be bigger than the population number.")
        print("Your population number: " + str(n))
        print("Your total space area: " + str(enclosed_square))
    else: 
        break

log_file = open("Log_" + str(file_number) + ".txt", "w+")
write_log(40 * '-')
write_log("Here's the Input Data:")
write_log(40 * '-') 
write_log('Total population number:', n)
write_log('Horizontal length:', xlimit)
write_log('Vertical length:', ylimit)
write_log('Total space is:', enclosed_square, "feet squared.")
write_log('Max distance (in feet) required for contamination:', Distlimit)

print("\n\nProcessing Simulation...")

# Day = 0
Day = 0 
df, Stat, MoversList = Generate(n, xlimit, ylimit) # This line only needs to be declared once.
infect(random.randrange(n))
fig, axs = plt.subplots(2)
fig.suptitle('Covid-19 Simulation', fontsize = 16)
Plot()
Count(Day)
write_log(40 * '-')
write_log('Day:',Day)
write_log(40 * '-')    
write_log(Stat.loc[Day])


# Day = 1
YesterdayPatients = list(df['Covid-19'])
Tomorrow()
infect(random.randrange(n)) # Will randomly infect people who are within a specific feet from another infected person
fig, axs = plt.subplots(2) # Properly labels the graphs that will be generated 
fig.suptitle('Covid-19 Simulation', fontsize = 16) # Gives the graph a title on top.
Plot()
Count(Day)
write_log(40 * '-')
write_log('Day:',Day)
write_log(40 * '-')    
write_log(Stat.loc[Day])
    
# Day 2 and 3 and so on 
same_result = 0
# While loop that will simulate COVID transmission as each day passes.
#while Stat.loc[Day, 'Not Infected'] > 0 and Day < 100:
#while Day < 100:
while Day < 100:
    log_file = open("Log_" + str(file_number) + ".txt", "a+")
    if (Stat.loc[Day, 'Infected'] == 0 and Stat.loc[Day, 'Hospitalized'] == 0):
        break
    else :
        same_result = 0
    
    YesterdayPatients = list(df['Covid-19'])
    Tomorrow()
    infect(random.randrange(n))
    fig, axs = plt.subplots(2)
    fig.suptitle('Covid-19 Simulation', fontsize = 16)
    Plot()
    Count(Day)
    write_log(40 * '-')
    write_log('Day:',Day)
    write_log(40 * '-')    
    write_log(Stat.loc[Day])
    log_file.close()
    

create_gif() 
print("\nSimulation Completed!")
Stat.to_excel('COVID-Stat_' + str(n) + '.xlsx')
Stat.plot(title = 'Statistical Data Vs. Days Passed')
plt.savefig('COVID-Stat-Graph_' + str(n)) 
```
